```python
import cv2
import os

def extract_frames(video_path, output_path, fps, video_index):
    # Create a VideoCapture object
    cap = cv2.VideoCapture(video_path)

    # Check if camera opened successfully
    if not cap.isOpened():
        print(f"Error opening video file: {video_path}")
        return

    # Get the total number of frames in the video
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    # Get the actual fps of the video
    video_fps = cap.get(cv2.CAP_PROP_FPS)
    # Calculate the frame interval
    frame_interval = int(video_fps / fps)

    # Read until video is completed
    frame_id = 0
    saved_frame_id = 0
    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()
        if not ret:
            break

        # Save each frame into a file if it's time
        if frame_id % frame_interval == 0:
            name = os.path.join(output_path, f'video{video_index}_frame{saved_frame_id}.png')
            cv2.imwrite(name, frame)
            saved_frame_id += 1

        # Increase frame id
        frame_id += 1

    # When everything done, release the video capture object
    cap.release()
    # Close all the frames
    cv2.destroyAllWindows()

# List of video paths
video_paths = [ "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\_import_6295b5ab77aa30.71180634.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\115-Lambafikira (92)_FPpreview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1832094_glasses_expression face_guy_1920x1080 (1).mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\2013058_bus_black_casual_1920x1080.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\4894625_young_college_cute_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\4898993_Middle aged_father_uncle_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\4910523_Handsome_young_man_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\4944876_glasses_man_sick_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\4954764_laptop_man_computer_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\5071002_glasses_man_sick_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\5071004_tired_man_sleepy_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\5119833_man_young_people_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\5219915_young_cute_handsome_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\5220009_young_cute_handsome_3840x2160.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1015846693.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1072195262-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1084332583-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1090790177-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1093653377-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1094117519-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1098141015-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1103034617-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1103412081-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1103456017-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1104709163-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1104808409-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1105603393-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1106048061-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1106348279-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1107034373-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1109113661-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1109758755-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1109940059-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1109940075-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1109940139-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1109991223-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1110184447-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1110258837-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\1110487985-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3392758483-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3405667925-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3409152557-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3412054015-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3415232967-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3415791795-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3439599859-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3497430617-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3500854575-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3509202251-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3509206165-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3511103627-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3511104679-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3511109731-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\3515810153-preview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\h23af9f5e_PRKLRSS15_FPpreview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\h98882086_A028_05301344_C006_V1-0021.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\hc744398e_JSGSSKS44_FPpreview.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\istockphoto-908310120-640_adpp_is.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\istockphoto-918015586-640_adpp_is.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\istockphoto-1203316867-640_adpp_is.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\istockphoto-1254376272-640_adpp_is.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\istockphoto-1270639920-640_adpp_is.mp4",
    "C:\\Users\\Kalivemula Rajendra\\Desktop\\rajendra\\data1\\istockphoto-1270640491-640_adpp_is.mp4"]
# Output path where the frames will be saved
output_path = r"C:\Users\Kalivemula Rajendra\OneDrive - ICFAI University\output_frames"

# Make sure the output directory exists
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Frames per second to extract
fps = 5

# Process each video
for index, video_path in enumerate(video_paths):
    extract_frames(video_path, output_path, fps, index)
```
